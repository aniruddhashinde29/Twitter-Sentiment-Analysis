```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
tweet_data = pd.read_csv("us_election_tweets.csv")
del tweet_data['Unnamed: 0']
tweet_data = tweet_data.dropna(subset = ['tweet'])
tweet_data['tweet'] = tweet_data['tweet'].astype('string')
tweet_data.info()
```

```python
def lower_tweet(tw):
        return tw.lower()

tweet_data['tweet'] = tweet_data['tweet'].apply(lower_tweet)
tweet_data['tweet']
```

```python
cleaned_tweets = []
def clean_data(tw):
    punc = '''!()-[]{};:'"\‚Äú‚Äô,<>./?@$%^&*_~'''
    for c in tw:
        if c == '#':
            tw = tw.replace(c, ' ')
        elif c in punc:
            tw = tw.replace(c, '')
    return(tw)

tweet_data['tweet'] = tweet_data['tweet'].apply(clean_data)
cleaned_tweets = tweet_data['tweet']
cleaned_tweets
```

```python
stop_words_df = pd.read_csv("stop_words.txt", header = None)
stop_words = stop_words_df[0].tolist()
temp = []
stopWords_removed_twlist = []
def remove_stop_words(tw):
    for tweet in tw:
        tweet = tweet.split()
        for item in tweet:
            if "‚Äô" in item: continue
            if item not in stop_words:
                temp.append(item)
        stopWords_removed_str = ' '.join(temp) 
        temp.clear()
        stopWords_removed_twlist.append(stopWords_removed_str)
    return(stopWords_removed_twlist)

stopWords_removed_twlist = remove_stop_words(cleaned_tweets)
stopWords_removed_twlist
```

```python
word_count = {}
master_word_count = {}
def bag_of_words(tw):
    word_count.clear()
    tw = tw.split()
    for word in set(tw):
        word_count[word] = tw.count(word)
    return word_count

for i in range(0, len(stopWords_removed_twlist)):
    master_word_count[i] = {}
    master_word_count[i].update(bag_of_words(stopWords_removed_twlist[i]))

master_word_count
```

```python
import re
relation = ''
relation_list = []
def candidate_relation(tw):
    if bool(re.search("(?=.*(donald|trump))(?=.*(joe|biden))", tw)) == True:
              relation = "TB"
              return(relation)
    elif bool(re.search("trump|donald", tw)) == True:
        relation = "T"
        return(relation)
    elif bool(re.search("biden|joe", tw)) == True:
        relation = "B"
        return(relation)
    
    else:
                 relation = "None"
                 return(relation)

for tweet in stopWords_removed_twlist:
    relation_list.append(candidate_relation(tweet))
relation_list
```

```python
corpus_df = pd.read_csv("corpus.txt", header = None, delimiter = "\t")
corpus_df
```

```python
tweet_score_list = []
def tweet_score(tw):
    tw = tw.split()
    tscore = 0
    for word in tw:
        if corpus_df.isin([word]).any().any() == True:
            row_label = 0
            word_score = 0
            row_label = corpus_df[corpus_df[0] == word].index.item()
            word_score = corpus_df.at[row_label, 1]
            tscore = tscore + word_score
    return(tscore)

for tweet in stopWords_removed_twlist:
    tweet_score_list.append(tweet_score(tweet))

tweet_score_list
```

```python
#This code converts all raw scores from above into a scale of 0-1 with the lowest score being 0 and highest 1
scaled_score_list = []
minScore = min(tweet_score_list)
maxScore = max(tweet_score_list)
for score in tweet_score_list:
    value = (score - minScore) / (maxScore - minScore)
    scaled_score_list.append(value)

scaled_score_list
```
